```python
import seaborn  as sns
import pandas as pd
import numpy as np
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
iris.columns
```

```python
sns.regplot(x="sepal_length", y='petal_length', data=iris)
```

```python
sns.distplot(iris.sepal_length, kde=True)
```

```python
iris.columns
```

```python
sns.heatmap(iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
```

```python
sns.stripplot(x='species' , y='petal_length',data=iris)
```

```python
sns.swarmplot(x='species', y='petal_length', hue='species', data=iris)
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.barplot(x='survived', y='sex', hue='class', data=titanic)
# bar plot to show the central tendency of values grouped by a category.
```

```python
sns.countplot(x='class', data=titanic)
```

```python
sns.pointplot(x='class', y='survived', hue='sex', data=titanic)
```

```python
sns.boxplot(data=iris)
# box plot to show the distribution and identify outliers
```

```python
sns.pairplot(iris, hue='species')
# grid of scatter plots and histograms for pairwise relationships between numeric
#columns in a DataFrame
```

```python
sns.jointplot(data=iris, x='sepal_length', y='petal_length', hue='species', kind='scatter')
# combined scatter plot and histogram or KDE plot to show relationships between
#two variables
```

```python
sns.relplot(x='sepal_length', y='petal_length', kind='scatter', data=iris)
# Provides a high-level interface for creating relational plots (like scatter or line plots)
```

```python
sns.lmplot(x='sepal_length', y='petal_length', data=iris)
#Creates scatter plots with regression lines
```

```python
sns.ecdfplot(data=iris['petal_length'])
# n Empirical Cumulative Distribution Function (ECDF) plot
```

```python
sns.violinplot(x='sepal_width' , y='petal_width', data=iris)
# Combines a box plot and KDE to show the distribution of data
```

```python
sns.swarmplot(x='sepal_width', y='petal_width', data=iris)
#swarmplot for categorical data to show individual data points
```

```python
sns.kdeplot(x='sepal_width', y='petal_width', data=iris)
#Kernel Density Estimate plot to visualize the probability density function of a
#dataset
```
